```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
education = pd.read_csv('educational-attainment-for-the-population-25-years-and-over.csv')
employment_status = pd.read_csv('employment-status-for-the-population-16-years-and-over.csv')
race = pd.read_csv('race.csv')
school_enrollment = pd.read_csv('school-enrollment-by-detailed-level-of-school-for-the-population-3-years-and-over.csv')
vacancy_status = pd.read_csv('vacancy-status.csv')
income = pd.read_csv('wage-or-salary-income-in-the-past-12-months-for-households.csv')
house_value = pd.read_csv('value.csv')
```

```python
education
```

```python
plt.figure(figsize=(20,10))
plt.plot(education['Neighborhood'],education['Estimate; Total: - No schooling completed'])
plt.grid()
plt.ylabel('Total')
plt.xlabel('Neighborhood')
#plt.savefig('Time_series.png')
plt.show()
```

```python
xi = education['Estimate; Total: - No schooling completed'].sort_values(ascending=False)[:30]
yi = education.sort_values(by=['Estimate; Total: - No schooling completed'],ascending=False).Neighborhood[:30]

plt.style.use('fivethirtyeight')
plt.subplots(figsize=(20,15))
sns.barplot(x=xi,y=yi,color='r')

sns.set(font_scale=3)   
sns.despine(left=True, bottom=True)

sns.set(font_scale=2)
plt.xlim=(0,25)
plt.ylabel("Neighborhood")
plt.xlabel("No schooling completed")
plt.title('')
plt.show()
```
